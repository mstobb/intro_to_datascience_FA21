# On Data Aesthetics

Data visualizations are **more** than just data.  They are also *art*!  The particular medium happens to be charts and graphs, but it is just as concerned with color, shape, size, and form.  


Art can have many different goals:
- Evoke an emotion
- Create beauty
- Understand a new concept

Data visualization also has a clear goal: **to clearly communicate true information.**

### Bad Art

Perhaps no art is *universally* bad, but there is art that we like and art that we may not like.  To better understand **why**, we need to develop a vocabulary that allows us to compare artist pieces.  

An **aesthetic** is

> a set of principles underlying and guiding the work of a particular artist or artistic movement

We need to develop an aesthetic for data visualizations just like we might for any other form of art.

## Edward Tufte

An artist and statistician, Edward Tufte thought a lot about how charts and graphs are, or should be, thought of as art.  More importantly, he thought about what aspects of the medium might help define a *good* chart from a *bad* chart.  

He wrote a book titled *The Visual Display of Quantitative Information*, which has become a classic in the data science field.

![Tufte Book](https://images-na.ssl-images-amazon.com/images/I/41IXJO4Mk6L.jpg)

The aesthetic that Tufte develops can generally be summarized in the following principles:

1. Maximize the **Data** to **Ink** ratio.
2. Minimize the **visual lie**.
3. Minimize **chartjunk**.
4. Use accurate **scales** and **labels**.
5. Use **color** effectively.
6. Exploit **repetition**.

## Minimize the Visual Lie

Remember, our goal is to clearly communicate **true information**.  However, it is altogether too easy to mislead with a visual image, while still keeping the raw data unchanged.

Generally, we will seek to match the size of an effect in the data with the size of the effect when we visualize it.  Some of the bad practices that fail to do this include the following:

- **Means without Variances**  
$x_1$ = [100, 100, 100, 100]  
$x_2$ = [200,   0, 10, 190]  
These two data sets have the same mean, but VERY different variances.  If we only reported the means, they would "look" the same, even though they are *very* different.

- **Showing interpolations without data**  
Interpolations of data smooth away the actual points and can tell a different, simplified story.

- **Distorting the Scales**  
The scales of the axes in plots can be scaled different based on the need of the figure.  These scales should be carefully selected to not convey a false representation of the data.


Crazy scales of market data

- **Removing the Scales Entirely**  
The only way we can understand the numerical values is with a scale of some type.  Removing them removes our ability to understand the data.

## Using Accurate Scales and Labels

Having incorrect or poorly made scales and labels is likely the single most common reason why charts and graphs can be misleading.  Unfortunately, there is no "perfect" scaling or labeling method.  Everyone needs to make the best decision possible at the time making the figure, so long as they are guided by main goal appropriately.

## Use Colors Effectively

Colors are great!  But they must be used correctly otherwise they hinder the graph more than they help.

Colors should mainly be used to
- distinguish between different classes
- visually encode numerical values
These cannot happen though if the colors cannot be distinguished from each other.  Using strongly distinct colors is thus important.

Best practices are to use colors such that they link with the object they are encoding.  Some examples might be:
- Financial losses in red
- Environmental spending in green
- Sport teams with their team colors

When using colors to encode number scales, such as from -1 to 1, much more attention is needed.  There are *MANY* ways to create a color scale.

## Exploit Repetition

Our brains are pattern recognition machines.  This can be utilitzed by repeating visual elements in graphics, so long as those elements represent the same type of object.  

- Use the same color and symbol for a class consisently
- Keep the axes the same scale between figures
- Complex plots can be repeated for different data sets
