## Goals of Sports Analytics
- Identify the type and frequency of data to be collected
- Create informative metrics by which to compare players and teams
- Predict the outcome of players or teams
- Interpret the results to real world actions


## Some of the Stats Used in Moneyball

#### On Base plus Slugging (OPS)
- On base percentage plus slugging percentage:  
  
$$ \frac{\text{Times On Base} + \text{Times Slugged}}{\text{Total At Bats}} $$  

- This one number gives us an idea of how often the player contributes to getting on base
- Different positions will have different OPS, but anything above 80% is considered good

#### Runs Created
- How many runs did the player contribute to the team:  
$$ \frac{(\text{Hits} + \text{Base on Ball}) \times \text{Total Bases}}{\text{At Bats} + \text{Base on Ball}} $$
- Gives a simple measurement of the offensive strength of a player

#### Weighted On Base Average (wOBA)
- An expansion of the standard batting average (hits/attempts) because not all hits are the same
- ![wOBA](https://wikimedia.org/api/rest_v1/media/math/render/svg/ff52e783994d7e2fb2b7bdf0d962f2ef6ab06e47)
- Where the different parts are defined as:
  - NIBB = Non-intentional bases on balls
  - HBP = Hit by pitch
  - 1B = Single
  - 2B = Double
  - 3B = Triple
  - HR = Home run
  - AB = at bat
  - BB = Base on balls
  - IBB = Intentional base on balls
  - SF = sacrifice flies
  - HBP = Hit by pitch

#### Fielding Independent Pitching (FIP)
- Earned Run Average (ERA) is a simple measure of how many runs a pitcher "allows"
- They would desire a low ERA
- But we desire to measure pitching ability completely independent of any fielder's influence  
  $$ \frac{13\text{Homeruns} + 3 \text{Base on Ball} - 2 \text{Strikeouts}}{\text{Innings Pitched}}  $$
- This is good at predicted the ERA of the pitcher in the following year



## Other Sports
- All sports are collecting data, but some have more challenges than others
- Sports like Baseball and Cricket operate in discrete chunks
  - Events generally only occur at a pitch
  - Most events end in few individual outcomes
- Most other sports happen over continuous time
  - Like Basketball, Soccer, and American Football
  - Now we have to track the interaction of all players and their movement on the field


## Future of Sports Analytics
- In many ways, the use of data in sports is **way ahead** of other industries
- This is driven by a few factors:
  - Extreme competition (literally!)
  - Generally easy to collect data
  - Generally easy to create informative metrics
- However, most data collected is of the physical variety
- Future sports analytics will seek to better understand and incorporate the mental aspects of sports
